### Pr√©traitement des donn√©es

Nous allons effectuer certaines transformations sur toutes les images (aussi sur les images de r√©f√©rence) d'un ou plusieurs ensembles de donn√©es, puis enregistrer les r√©sultats dans un r√©pertoire cible. Ce faisant, le g√©n√©rateur reproduit la m√™me hi√©rarchie de r√©pertoires que le r√©pertoire source dans le r√©pertoire cible.

Si des donn√©es de v√©rit√© terrain sont fournies, elles sont trait√©es de mani√®re coh√©rente avec les images respectives.

#### Chemins d'acc√®s √† la source et √† la destination

Nous allons :

- D√©finir les chemins d'acc√®s √† un nombre arbitraire de jeux de donn√©es sources que l'on souhaite traiter.
- D√©finir un r√©pertoire d'exportation, c'est-√†-dire le r√©pertoire dans lequel on souhaite exporter les donn√©es trait√©es.
- D√©finir le r√©pertoire racine des jeux de donn√©es √† traiter. Le g√©n√©rateur reproduira la hi√©rarchie des r√©pertoires qu'il trouve dans ce r√©pertoire racine dans le r√©pertoire d'exportation.

```python
# Les r√©pertoires qui contiennent les images train, val et test
train_images = 'data/cityscapes/leftImg8bit/train/'
val_images = 'data/cityscapes/leftImg8bit/val/'
test_images = 'data/cityscapes/leftImg8bit/test/'

# Les r√©pertoires qui contiennent les images de v√©rit√© terrain (train et val)
train_gt = 'data/cityscapes/gtFine/train/'
val_gt = 'data/cityscapes/gtFine/val/'

# D√©finit les √©l√©ments ci-dessus √† passer au constructeur du `BatchDataGenAug`.
image_dirs = [train_images, val_images]
ground_truth_dirs = [train_gt, val_gt]

# Le r√©pertoire dans lequel les images trait√©es doivent √™tre export√©es.
export_dir = 'data/cityscapes_small/'
# Le r√©pertoire racine du jeu de donn√©es.
root_dir = 'data/cityscapes/'
```

#### Instanciation d'une pipeline du g√©n√©rateur de donn√©es

Nous commen√ßons donc par cr√©er une instance de BatchDataGenAug et nous lui passons une liste de r√©pertoires contenant les images avec les images de v√©rit√© terrain que nous souhaitons traiter.

```python
offline_processor = BatchDataGenAug(image_dirs=image_dirs,
                                    image_file_extension='png',
                                    ground_truth_dirs=ground_truth_dirs,
                                    image_name_split_separator='leftImg8bit',
                                    ground_truth_suffix='gtFine_labelIds',
                                    check_existence=True,
                                    root_dir=root_dir,
                                    export_dir=export_dir)

num_files = offline_processor.get_num_files()
print("Nombre total de fichiers dans tous les jeux de donn√©es :", num_files)
```

#### R√©glage des param√®tres de traitement et d√©marrage du traitement

Il suffit d'appeler ***process_all()*** sur l'instance de **BatchGenerator** avec les param√®tres souhait√©s.

Dans cet exemple, nous souhaitons faire deux choses avec les donn√©es :

1. Les ID des classes de segmentation dans les images de v√©rit√© de terrain Cityscapes vont de 0 √† 33 et contiennent un grand nombre de classes non pertinentes qui sont ignor√©es dans l'√©valuation. Donc, nous allons transmettre une carte sous la forme d'un dictionnaire pour en conserver les 8 principales, cela va ralentir le traitement. Lorsque le mod√®le a √©t√© form√©, il est possible de r√©tablir les ID de classe d'origine de la m√™me mani√®re.

2. Les images de l'ensemble de donn√©es Cityscapes ont une taille de (1024, 2048). Il serait souhaitable que le mod√®le soit entrain√© sur ces images en taille r√©elle, mais cela est trop important pour le GPU. En cons√©quence, nous allons r√©duire la taille des images d'un facteur 16 √† (256, 512).

Consultez [la documentation du script](data/doc/batchdatagenaug.html) pour plus de d√©tails sur les autres transformations disponibles dans BatchDataGenAug.

```python
offline_processor.process_all(
    convert_colors_to_ids=False,
    # Convertit les identifiants de classe en d'autres.
    convert_ids_to_ids=IDS_TO_CATEGORYIDS_ARRAY,
    convert_to_one_hot=False,
    void_class_id=None,
    random_crop=False,
    crop=False,
    # Redimensionne les images
    resize=(256, 512),
    brightness=False,
    flip=False,
    translate=False,
    scale=False,
    gray=False,
    # Enregistre les images trait√©es sur le disque.
    to_disk=True)
```

### Cr√©er un g√©n√©rateur pour l'entrainement et l'√©valuation

Les mod√®les ont besoin d'un g√©n√©rateur qui les alimente avec des lots d'images et les images de r√©f√©rence (ground truth) correspondantes. Id√©alement, nous voulons deux g√©n√©rateurs, un qui sert les donn√©es d'un jeu de donn√©es d'entra√Ænement et un autre qui sert les donn√©es d'un jeu de donn√©es de validation. Le jeu de donn√©es Cityscapes fournit d√©j√† une partie des donn√©es pour nous, donc nous allons nous en tenir √† cela.

Afin de s'entra√Æner avec le jeu de donn√©es Cityscapes, la seule chose que nous devons vraiment faire ici est de d√©finir les chemins appropri√©s vers le jeu de donn√©es sur notre machine, pour d'autres jeux de donn√©es, vous devrez passer des valeurs diff√©rentes au constructeur `BatchDataGenAug`, consultez la documentation pour plus de d√©tails.

Note : Si vous devez pr√©traiter votre jeu de donn√©es, par exemple pour changer la taille de l'image ou pour convertir les √©tiquettes de classe de segmentation, je vous sugg√®re de le faire hors ligne.

#### Donn√©es d‚Äôentrainement, de test et de validation

```python
# D√©finir les chemins vers les images.
train_images = 'data/cityscapes_small/leftImg8bit/train/'
val_images = 'data/cityscapes_small/leftImg8bit/val/'
test_images = 'data/cityscapes_small/leftImg8bit/test/'

# D√©finir les chemins vers les images ground truth.
train_gt = 'data/cityscapes_small/gtFine/train/'
val_gt = 'data/cityscapes_small/gtFine/val/'

# Mettre les chemins vers les jeux de donn√©es dans des listes, parce que
# c'est ce que `BatchDataGenAug` exige comme entr√©e.
train_image_dirs = [train_images]
train_ground_truth_dirs = [train_gt]
val_image_dirs = [val_images]
val_ground_truth_dirs = [val_gt]
```

```python
# D√©finir le nombre de classes de segmentation.
num_classes = 8

train_dataset = BatchDataGenAug(image_dirs=train_image_dirs,
                                image_file_extension='png',
                                ground_truth_dirs=train_ground_truth_dirs,
                                image_name_split_separator='leftImg8bit',
                                ground_truth_suffix='gtFine_labelIds',
                                check_existence=True,
                                num_classes=num_classes)

val_dataset = BatchDataGenAug(image_dirs=val_image_dirs,
                              image_file_extension='png',
                              ground_truth_dirs=val_ground_truth_dirs,
                              image_name_split_separator='leftImg8bit',
                              ground_truth_suffix='gtFine_labelIds',
                              check_existence=True,
                              num_classes=num_classes)

num_train_images = train_dataset.get_num_files()
num_val_images = val_dataset.get_num_files()

print("Taille du jeu de donn√©es d'entra√Ænement :", num_train_images, "images")
print("Taille du jeu de donn√©es de validation :", num_val_images, "images")
```

### Entra√Ænement du mod√®le

Maintenant, il suffit d'appeler la m√©thode `train()` pour entra√Æner le mod√®le. R√©f√©rez-vous √† la documentation pour plus de d√©tails sur tous les arguments, mais voici quelques notes :

1. Vous devrez passer une fonction de planification du taux d'apprentissage, m√™me si elle est simple. Cette fonction prend en entr√©e un entier (le pas d'apprentissage) et retourne un flottant (le taux d'apprentissage). Je vais simplement d√©finir une fonction de pas simple ci-dessous.
2. Passez le(s) g√©n√©rateur(s) que nous avons instanci√©(s) ci-dessus. Notez qu'il y a deux arguments qui prennent un g√©n√©rateur en entr√©e, `train_generator` et `val_generator`, o√π le dernier est optionnel.

```python
epochs = 6  # D√©finir le nombre d'√©poques pour l'entra√Ænement.


# D√©finir une fonction de planification du taux d'apprentissage √† passer
# √† la m√©thode `train()`.
def learning_rate_schedule(step):
    if step <= 10000:
        return 0.0001
    elif 10000 < step <= 20000:
        return 0.00001
    elif 20000 < step <= 40000:
        return 0.000003
    else:
        return 0.000001


model.train(train_generator=train_generator,
            epochs=epochs,
            steps_per_epoch=ceil(num_train_images / batch_size),
            learning_rate_schedule=learning_rate_schedule,
            keep_prob=0.5,
            l2_regularization=0.0,
            eval_dataset='val',
            eval_frequency=2,
            val_generator=val_generator,
            val_steps=ceil(num_val_images / batch_size),
            metrics={'loss', 'mean_iou', 'accuracy'},
            save_during_training=True,
            save_dir='cityscapes_model',
            save_best_only=True,
            save_tags=['default'],
            save_name='(batch-size-4)',
            save_frequency=2,
            saver='saved_model',
            monitor='loss',
            record_summaries=True,
            summaries_frequency=10,
            summaries_dir='data/logs',
            summaries_name='configuration_01',
            training_loss_display_averaging=3)
```

### U-Net

<img src="data/images/U-Net.png" width='500'>

Le r√©seau U-net s'appuie sur le FCN d√©crit ci-dessus. Il se compose √©galement d'un encodeur qui r√©duit l'√©chantillonnage de l'image d'entr√©e √† une carte de caract√©ristiques et d'un d√©codeur qui augmente l'√©chantillonnage de la carte de caract√©ristiques √† la taille de l'image d'initial en utilisant des couches de d√©convolution apprises.

La principale contribution de l'architecture U-Net r√©side dans les connexions de raccourci. Nous avons vu plus haut dans FCN que puisque nous sous-√©chantillonnons une image dans le cadre de l'encodeur, nous avons perdu beaucoup d'informations qui ne peuvent pas √™tre facilement r√©cup√©r√©es dans la partie encodeur. FCN tente de r√©soudre ce probl√®me en prenant des informations dans les couches de mise en commun avant la couche de caract√©ristiques finale.

U-Net propose une nouvelle approche pour r√©soudre ce probl√®me de perte d'information. Il propose d'envoyer des informations √† chaque couche d'√©chantillonnage ascendante dans le d√©codeur √† partir de la couche d'√©chantillonnage descendante correspondante dans le d√©codeur, comme on peut le voir dans l'illustration ci-dessus. Puisque les couches au d√©but de l'encodeur auraient plus d'informations, elles soutiendraient l'op√©ration d'√©chantillonnage ascendant du d√©codeur en fournissant des d√©tails fins correspondant aux images d'entr√©e, ce qui am√©liorerait consid√©rablement les r√©sultats.

(source : https://arxiv.org/abs/1505.04597)

### SegNet

<img src="data/images/segnet.png" width='500'>

Alors que dans un mod√®le U-Net, des cartes de caract√©ristiques enti√®res sont transf√©r√©es du chemin de compression au chemin d'expansion, ce qui n√©cessite beaucoup de m√©moire. Dans Segnet, seuls les indice de pooling sont transf√©r√©s vers le chemin d'expansion √† partir du chemin de compression, utilisant moins de m√©moire. 

SegNet est compos√© d'un r√©seau encodeur et d'un r√©seau d√©codeur correspondant, suivis d'une couche de classification finale par pixel. Cette architecture est illustr√©e dans le sch√©ma ci-dessus. Le r√©seau encodeur est compos√© de 13 couches convolutionnelles qui correspondent aux 13 premi√®res couches convolutionnelles du r√©seau VGG16 con√ßu pour la classification d'objets. Ils √©cartent les couches fully connected en faveur de la r√©tention de cartes de caract√©ristiques de plus haute r√©solution √† la sortie la plus profonde de l'encodeur. Cela permet √©galement de r√©duire consid√©rablement le nombre de param√®tres dans le r√©seau encodeur SegNet (de 134M √† 14,7M) par rapport √† d'autres architectures r√©centes. √Ä chaque couche d'encodage correspond une couche de d√©codage et le r√©seau de d√©codage compte donc 13 couches. La sortie finale du d√©codeur est transmise √† un classificateur soft-max multi-classes pour produire des probabilit√©s de classe pour chaque pixel ind√©pendamment.
